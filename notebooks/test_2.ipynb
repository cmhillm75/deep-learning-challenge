## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
# Read in the charity_data.csv. 
# ("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
csv_path = "/app/data/raw/charity_data.csv"
application_df = pd.read_csv(csv_path)

# Display the DataFrame
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()

# Display the unique value counts for each column
print(unique_values)
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Display the value counts 
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# Set the cutoff value to 200 and create the list
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[
    application_df['APPLICATION_TYPE'].value_counts() < 550
].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_type_counts = application_df['CLASSIFICATION'].value_counts()
classification_type_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# Show only classification_type_counts >= 1
# Temporarily show all rows for classification_type_counts >= 1
with pd.option_context('display.max_rows', None):
    print(classification_type_counts[classification_type_counts >= 1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Set the cutoff value to 1000 and create the list
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[
    application_df['CLASSIFICATION'].value_counts() < 1000
].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Create bins/labels for the "ASK_AMT" to optimize the input variables

# Define bins and labels
bins = [0, 5000, 10000, 20000, 50000, 100000, 500000, 1000000, float('inf')]
labels = ['<5k', '5k-10k', '10k-20k', '20k-50k', '50k-100k', '100k-500k', '500k-1M', '>1M']

# Create a new binned column
application_df['ASK_AMT_BINS'] = pd.cut(application_df['ASK_AMT'], bins=bins, labels=labels)

# Check the new column to verify
application_df['ASK_AMT_BINS'].value_counts()
```

```python
# Confirm data types
application_df.dtypes
```

```python
# Drop the raw ASK_AMT column
application_df = application_df.drop(columns=['ASK_AMT'])

# Confirm the column was removed
application_df.dtypes
```

```python
# Make a copy of application_df
application_df_copy = application_df.copy()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
category_columns = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 
                    'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS', 'ASK_AMT_BINS']

application_df_copy = pd.get_dummies(application_df, columns=category_columns)

# Show updated dummies df
application_df_copy.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_copy.drop(columns=['IS_SUCCESSFUL'])
y = application_df_copy['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Changed units from 1st attempt and added in Dropout element to randomize deactivation of 20% of neurons
def create_model():
    nn = tf.keras.models.Sequential()
    # First hidden layer with Dropout
    nn.add(tf.keras.layers.Dense(units=100, activation='relu', input_dim=X_train_scaled.shape[1]))
    nn.add(tf.keras.layers.Dropout(0.2))  

    # Second hidden layer with Dropout
    nn.add(tf.keras.layers.Dense(units=50, activation='relu'))
    nn.add(tf.keras.layers.Dropout(0.2))  

    # Output layer
    nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
    return nn

# Create the model
nn_model = create_model()

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model aka Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, 
                     epochs=150, batch_size=32, 
                     validation_split=0.2)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save('/app/data/processed/AlphabetSoupCharity_Optimization.h5')
```
